```python
import numpy as nm
import pandas as pn
```

```python
col_name = ['CA1','CA2','CA3']
sit_no = nm.arange(1,31)
nm.random
score = nm.random.randint(1,11,90)
score = score.reshape(30,3)
student_score = pn.DataFrame(data = score, index=sit_no, columns=col_name)
student_score
```

```python
nm.random
student_score['ExamsScore'] = nm.random.randint(25,71,30).reshape(30,1)
student_score
```

```python
nm.random
student_score.insert(loc=0, value=nm.random.randint(20,28,30), column='AGE')
student_score
```

```python
nm.random
student_score.insert(loc=4, value=nm.random.randint(1,11,30),column='CA4')
student_score
```

```python
student_score['TotalScore'] = student_score['CA1'] + student_score['CA2'] + student_score['CA3'] + student_score['ExamsScore']
student_score
```

```python
student_score.drop('CA4', axis=1, inplace=True) # dropping the column CA4 from the dataFrame using its name(CA4), axis=1(column) and inplace as True
student_score
```

```python
def grade(TotalScore):
    if TotalScore >= 80 and TotalScore <=100:
        return 'A'
    elif TotalScore >= 70 and TotalScore <79.99:
        return 'B'
    elif TotalScore >= 60 and TotalScore <69.99:
        return 'C'
    elif TotalScore >= 50 and TotalScore <59.99:
        return 'D'
    elif TotalScore >= 40 and TotalScore <49.99:
        return 'E'
    else:
        return 'F'
student_score['GRADE'] = student_score['TotalScore'].apply(grade)
student_score
```
